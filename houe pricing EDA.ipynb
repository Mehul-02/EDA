```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
pd.pandas.set_option('display.max_columns',None)
```

```python
df = pd.read_csv(r'D:\mehul sirohi\train.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
features_with_na=[features for features in df.columns if df[features].isnull().sum()>1]
```

```python
features_with_na
```

```python
for features in features_with_na:
    print(features,np.round(df[features].isnull().mean()*100,4), '%missing values')
```

```python
plt.figure(dpi = 150)
sns.boxplot(df['SalePrice'])
```

```python
data=df.copy()
for feature in features_with_na:
    data[feature]=np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
df.info()
```

```python
df['SalePrice'].dtype!='O'
```

```python
numerical_features= [feature for feature in df.columns if df[feature].dtypes!='O']
```

```python
print(len(numerical_features))
df[numerical_features].head()
```

```python
year_feature=[feature for feature in numerical_features if'Yr' in feature or 'Year' in feature]
```

```python
for feature in year_feature:
    print(feature,df[feature].unique())
```

```python
plt.figure(dpi=150)
df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs Year Sold')
```

```python
data=df.copy()
for feature in year_feature:
    if feature!='YrSold':
        data[feature]=data['YrSold']-data[feature]
        plt.figure(dpi=150)
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
decrete_features = [feature for feature in numerical_features if len(df[feature].unique())<=25]
print(len(decrete_features))
```

```python
df[decrete_features].head()
```

```python
data=df.copy()
for features in decrete_features:
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(features)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
continuous_features = [feature for feature in numerical_features if feature not in decrete_features+year_feature]
print(len(continuous_features))
```

```python
data=df.copy()
for feature in continuous_features:
    data.groupby(feature)['SalePrice'].median().plot.hist()
    plt.xlabel(feature)
    plt.ylabel('SalesPrice')
    plt.title(feature)
    plt.show()
```

```python
#we will be using logarithmic transformation
data=df.copy()
for feature in continuous_features:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()
```

```python
#outliers
for feature in continuous_features:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
        
```
